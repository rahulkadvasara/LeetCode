Given a string s and an integer k, return the length of the longest substring of s such that the frequency of each character in this substring is greater than or equal to k.

if no such substring exists, return 0.

 

Example 1:

Input: s = "aaabb", k = 3
Output: 3
Explanation: The longest substring is "aaa", as 'a' is repeated 3 times.
Example 2:

Input: s = "ababbc", k = 2
Output: 5
Explanation: The longest substring is "ababb", as 'a' is repeated 2 times and 'b' is repeated 3 times.
 

Constraints:

1 <= s.length <= 104
s consists of only lowercase English letters.
1 <= k <= 105

```python
class Solution:
    def longestSubstring(self, s: str, k: int) -> int:
        count = 0
        for i in range(1, 27):
            count = max(count, self.helper(s, k, i))
        return count

    def helper(self, s, k, numUniqueTarget):
        start = end = numUnique = numNoLessThanK = count = 0
        chMap = [0]*128

        while end < len(s):
            if chMap[ord(s[end])] == 0: numUnique += 1
            chMap[ord(s[end])] += 1
            if chMap[ord(s[end])] == k: numNoLessThanK += 1
            end += 1
            
            while numUnique > numUniqueTarget:
                if chMap[ord(s[start])] == k: numNoLessThanK -= 1
                chMap[ord(s[start])] -= 1
                if chMap[ord(s[start])] == 0: numUnique -= 1
                start += 1
                
            if numUnique == numNoLessThanK: count = max(count, end-start)
            
        return count
```

```python
class Solution:

    def longestSubstring(self, s: str, k: int) -> int:
        
        if len(s) < k:
            return 0
        
        freq = {}

        for char in s:
            freq[char] = freq.get(char, 0) + 1
        splits = []

        for i, char in enumerate(s):
            if freq[char] < k:
                splits.append(i)
    
        if not splits:
            return len(s)
    
        start = 0
        max_len = 0

        for split in splits:
            sub = s[start:split]
            if sub:  
                max_len = max(max_len, self.longestSubstring(sub, k))
            start = split + 1

        if start < len(s):
            sub = s[start:]
            max_len = max(max_len, self.longestSubstring(sub, k))
    
        return max_len
```

```python
class Solution:
    def longestSubstring(self, s: str, k: int) -> int:
        # Example: s = "abcdef", k = 3; len = 6 (0, 1, 2, 3, 4, 5)
        # * Brute force: O(N^2 * N)
        # * More optimal: O(N^2)

        # Helper function:
        def check(letters_dict) -> bool:
            # return all([count >= k for letter, count in letters_dict.items()])
            for letter, count in letters_dict.items():
                if count < k:
                    return False
            return True

        # Constant:
        LEN_S = len(s)

        answer = 0
        for left in range(LEN_S - k + 1):
            # Reset and initialize the dictionry.
            letters_dict = defaultdict(int)

            # Extend the substring.
            for right in range(left, LEN_S):
                # Update the dictionary.
                letters_dict[s[right]] += 1

                if check(letters_dict):
                    answer = max(answer, right - left + 1)
        return answer
```
