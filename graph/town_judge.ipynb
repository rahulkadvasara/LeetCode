997. Find the Town Judge
Easy

In a town, there are n people labeled from 1 to n. There is a rumor that one of these people is secretly the town judge.
If the town judge exists, then:
The town judge trusts nobody.
Everybody (except for the town judge) trusts the town judge.
There is exactly one person that satisfies properties 1 and 2.
You are given an array trust where trust[i] = [ai, bi] representing that the person labeled ai trusts the person labeled bi. If a trust relationship does not exist in trust array, then such a trust relationship does not exist.
Return the label of the town judge if the town judge exists and can be identified, or return -1 otherwise.

Example 1:
Input: n = 2, trust = [[1,2]]
Output: 2

Example 2:
Input: n = 3, trust = [[1,3],[2,3]]
Output: 3

Example 3:
Input: n = 3, trust = [[1,3],[2,3],[3,1]]
Output: -1

Constraints:
1 <= n <= 1000
0 <= trust.length <= 104
trust[i].length == 2
All the pairs of trust are unique.
ai != bi
1 <= ai, bi <= n

```python
class Solution:
    def findJudge(self, n: int, trust: List[List[int]]) -> int:
        in_degree = [0] * (n + 1)
        out_degree = [0] * (n + 1)
        for a in trust:
            out_degree[a[0]] += 1
            in_degree[a[1]] += 1
        for i in range(1, n + 1):
            if in_degree[i] == n - 1 and out_degree[i] == 0:
                return i
        return -1
```

```python
from collections import defaultdict
class Solution:
    def findJudge(self, n: int, trust: List[List[int]]) -> int:
        graph = defaultdict(list)
        for i, j in trust:
            graph[i].append(j)
        
        jlst = []
        for i in range(1, n+1):
            if graph[i] == []:
                jlst.append(i)
        if jlst == []:
            return -1
        
        for j in jlst:
            present = True
            for p in range(1, n+1):
                if p == j:
                    continue
                if j not in graph[p]:
                    present = False
                    break
            if present:
                return j
        return -1
```

```python
# Graph-based solution

from collections import defaultdict
class Solution:
    def findJudge(self, n: int, trust: List[List[int]]) -> int:
        # Build adjacency list for trust relationships
        adj = defaultdict(set)
        for a, b in trust:
            adj[a].add(b)
        
        candidates = []
        for person in range(1, n + 1):
            if len(adj[person]) == 0:
            candidates.append(person)
        
        for judge in candidates:
            if all(judge in adj[other] for other in range(1, n + 1) if other != judge):
            return judge
        return -1
```

```python
from collections import defaultdict
class Solution:
    def findJudge(self, n: int, trust: List[List[int]]) -> int:
        adj = defaultdict(list)
        for a, b in trust:
            adj[a].append(b)
        
        candidates = []
        for person in range(1, n + 1):
            if len(adj[person]) == 0:
                candidates.append(person)
        
        for judge in candidates:
            seen = set()
            for other in range(1, n + 1):
                if other != judge and judge in adj[other]:
                    seen.add(other)
            if len(seen) == n - 1:
                return judge
        return -1
```

```python
"""
The town judge trusts nobody → no outgoing edges.

Everyone else trusts the judge → n - 1 incoming edges to the judge.

These can be modeled and checked using in-degree and out-degree, classic graph concepts.

Modeling it as a graph simplifies the logic:
Out-degree = 0 for the judge
In-degree = n - 1 for the judge


Time complexity:
n is the number of people
t is the length of the trust list

Breakdown:
Initializing in_deg  → O(n)

Processing all trust relationships → O(t)

Scanning for the judge → O(n)

Total: O(n + t)


"""


class Solution:
    def findJudge(self, n: int, trust: List[List[int]]) -> int:

        in_deg=[0]*(n)
        #out_deg=[0]*(n)

        for a,b in trust:
            #out_deg[a-1]+=1
            in_deg[a-1]-=1 #Person a trusts someone → disqualifies them from being judge
            in_deg[b-1]+=1  # Person b is trusted → moves toward being judge
        
        for i in range(n):
            if in_deg[i]==n-1: #and out_deg[i]==0:
                return i+1
        
        return -1


        
```
